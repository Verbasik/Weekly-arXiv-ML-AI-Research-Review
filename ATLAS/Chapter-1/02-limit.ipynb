## Предел функции — финишная прямая

Давайте вернемся к нашему велосипедисту. Поездка подходит к концу, и он подъезжает к своему дому.

Представьте, что до дома остался всего один метр. Ваша скорость уже очень низкая. Когда осталось полметра — она еще ниже. Десять сантиметров... один сантиметр... один миллиметр... Вы всё ближе и ближе к точке остановки, и ваша скорость всё ближе и ближе к нулю.

**Предел** — это и есть то целевое значение, к которому стремится функция, когда её аргумент приближается к определённой точке.

В нашем примере:
- **Аргумент** — это расстояние до дома.
- **Функция** — это ваша скорость.

Чем ближе расстояние до дома к нулю, тем ближе ваша скорость к нулю. Даже если мы не рассматриваем сам момент остановки, мы можем с уверенностью сказать, к какому значению стремится скорость.

Математически это записывается так: «предел функции скорости при расстоянии, стремящемся к нулю, равен нулю».

### Предел и «сломанный» калькулятор

А теперь представим другую ситуацию, чтобы понять самую суть пределов. У вас есть калькулятор с одной единственной, но очень интересной функцией:

$f(x) = \frac{x^2 - 9}{x - 3}$

Этот калькулятор отлично работает для любого числа, кроме одного. Попробуйте ввести `x = 3`. Вы получите $\frac{9 - 9}{3 - 3} = \frac{0}{0}$. Это неопределенность, на ноль делить нельзя! Калькулятор выдаст ошибку.

Но нам любопытно: а какое число *должен был бы* показать калькулятор, если бы он не ломался? Какое значение «ожидается» в этой точке? Чтобы это выяснить, давайте не будем нажимать на «3», а будем «подкрадываться» к этому числу с разных сторон.

**1. Подходим к тройке «снизу» (числа чуть меньше 3):**

- `x = 2.9`  →  $f(2.9) = \frac{2.9^2 - 9}{2.9 - 3} = \frac{8.41 - 9}{-0.1} = \frac{-0.59}{-0.1} = 5.9$
- `x = 2.99` →  $f(2.99) = \frac{2.99^2 - 9}{2.99 - 3} = \frac{8.9401 - 9}{-0.01} = \frac{-0.0599}{-0.01} = 5.99$
- `x = 2.999` → $f(2.999) = 5.999$

Видите закономерность? Чем ближе мы к 3, тем ближе результат к **6**.

**2. Подходим к тройке «сверху» (числа чуть больше 3):**

- `x = 3.1`  →  $f(3.1) = \frac{3.1^2 - 9}{3.1 - 3} = \frac{9.61 - 9}{0.1} = \frac{0.61}{0.1} = 6.1$
- `x = 3.01` →  $f(3.01) = \frac{3.01^2 - 9}{3.01 - 3} = \frac{9.0601 - 9}{0.01} = \frac{0.0601}{0.01} = 6.01$
- `x = 3.001` → $f(3.001) = 6.001$

И снова та же картина! С какой бы стороны мы ни подходили к тройке, результат нашей функции неотвратимо стремится к числу **6**.

Это и есть предел. Мы можем сказать, что **предел функции $f(x)$ при $x$, стремящемся к 3, равен 6**.

### Запись и график

В математике для этого есть специальная запись:

$\lim_{x \to 3} \frac{x^2 - 9}{x - 3} = 6$

Расшифруем её:
- **lim** — сокращение от латинского *limes*, что и означает «предел».
- **$x \to 3$** — читается как «икс стремится к трём». Это показывает, к какому значению мы «подкрадываемся».
- **$\frac{x^2 - 9}{x - 3}$** — наша функция, поведение которой мы исследуем.
- **$= 6$** — то самое целевое значение, к которому стремится функция.

А как это выглядит на графике? Если мы упростим нашу функцию (используя формулу разности квадратов), то получим:
$f(x) = \frac{(x-3)(x+3)}{x-3} = x+3$

Это значит, что график нашей функции — это прямая линия $y = x+3$. Но с одним важным исключением: в точке `x = 3` наша исходная функция не существует!

На графике это будет выглядеть как идеально ровная прямая с одной «выколотой» или «пустой» точкой.



- По оси $X$ мы подходим всё ближе и ближе к `x = 3`.
- По оси $Y$ мы видим, что линия графика подходит всё ближе и ближе к `y = 6`.

**Предел — это и есть Y-координата этой «дырки» на графике.** Это то значение, которое идеально вписалось бы в это пустое место, чтобы сделать линию непрерывной.
