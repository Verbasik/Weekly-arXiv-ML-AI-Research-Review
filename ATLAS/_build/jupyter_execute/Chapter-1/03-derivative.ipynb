## Производная — узнаём скорость в любой момент

Мы уже поняли, как вычислить *среднюю* скорость нашего велосипедиста за какой-то промежуток времени. Мы брали две точки на его пути и делили изменение скорости на изменение времени ($\frac{\Delta y}{\Delta x}$).

Но это как сказать: «За последний час моя средняя скорость была 15 км/ч». Это не говорит вам, гнали ли вы с горы со скоростью 30 км/ч или еле тащились в гору со скоростью 5 км/ч.

А что, если мы хотим узнать скорость **прямо сейчас**? Как на спидометре. Как математически «посмотреть на спидометр» функции?

### Фокус с приближением

Для вычисления средней скорости нам нужны были две точки. А для мгновенной скорости у нас есть только одна — «сейчас». Вот тут и начинается самое интересное.

Давайте схитрим. Возьмем нашу точку «сейчас» (**A**) и вторую точку (**B**), но расположим её невероятно, *невообразимо* близко к первой.

Представьте, что вы смотрите на график движения. Сначала между точками **A** и **B** есть видимое расстояние. Но мы начинаем двигать точку **B** по графику всё ближе к **A**.

- Когда **B** находится в 1 секунде от **A**, мы получаем очень точную среднюю скорость за эту секунду.
- Когда **B** в 0.1 секунды от **A** — скорость еще точнее.
- Когда **B** в 0.000001 секунды от **A**...

Линия, которая соединяет **A** и **B** (секущая), почти перестает пересекать график, она почти просто *касается* его в одной точке.

Этот процесс «бесконечного приближения» вам уже знаком — это **предел**! Мы заставляем промежуток времени $\Delta x$ стремиться к нулю, чтобы поймать то самое значение скорости в одно-единственное мгновение.

### Что такое производная? Проще некуда!

**Производная** — это и есть та самая **мгновенная скорость** изменения функции.

> Если функция — это ваш путь, то производная — это ваш спидометр.

Она показывает, насколько быстро и в каком направлении меняется ваша функция (скорость, температура, прибыль — что угодно) в конкретной точке.

Обозначают производную очень просто — штрихом: **$f'(x)$**. Читается как «эф штрих от икс». Увидели штрих — знайте, речь идет о скорости изменения.

### Что производная показывает на графике?

Самое наглядное объяснение производной — через **касательную**.

**Касательная** — это прямая, которая касается графика функции в одной-единственной точке, как бы «прислоняется» к нему.

Так вот, **производная в точке — это наклон касательной в этой точке.**

*   Если касательная **смотрит вверх** — производная положительная. Это значит, что функция **растёт** (велосипедист разгоняется).
*   Если касательная **смотрит вниз** — производная отрицательная. Функция **убывает** (велосипедист тормозит).
*   Если касательная **горизонтальна** — производная равна нулю. Функция достигла пика («вершина холма») или дна («нижняя точка впадины»). Скорость в этот момент равна нулю.

**Итак, главное, что нужно запомнить:**

1.  **Производная — это мгновенная скорость функции.**
2.  **На графике она показывает наклон касательной.**
3.  **Она помогает понять, растет функция или убывает, и как быстро она это делает.**

Теперь, когда мы знаем, что такое производная, мы можем научиться находить её для разных функций, используя готовые правила, и это намного проще, чем вычислять пределы каждый раз.
